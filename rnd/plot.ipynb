```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
%matplotlib inline
import os
```

```python
images = []
imgdir = "/Users/ananth/Desktop/"
with open("/Users/ananth/Desktop/test_images.txt", 'r') as txt:
    for name in txt.readlines():
        name = imgdir + name.replace("\n", "")
        images.append(np.asarray(Image.open(os.path.join(imgdir, name), 'r')))

imgdir = "/Users/ananth/Documents/hackathon/drn/results/"
for name in filter(lambda x: ".png" in x, os.listdir(imgdir)):
    name = imgdir + name.replace("\n", "")
    images.append(np.asarray(Image.open(os.path.join(imgdir,name), 'r')))
```

```python
fig = plt.figure(figsize=(50, 15))  # width, height in inches

for i in range(6):
    sub = fig.add_subplot(2, 3, i + 1)
    sub.imshow(np.asarray(images[i]), interpolation='nearest')
```
